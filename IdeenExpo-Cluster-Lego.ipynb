```python
import numpy as np
from kmodes.kmodes import KModes
from kmodes.kprototypes import KPrototypes
import matplotlib.pyplot as plt
import matplotlib.cm as cm
```

```python
km = KModes(n_clusters=3, init='Huang', n_init=500,max_iter=500)
```

```python
data = np.genfromtxt('Lego-Parts-3.csv',dtype=None,delimiter=';',encoding=None)
```

```python
clusters = km.fit_predict(data)
```

```python
print(km.cluster_centroids_)
```

```python
plt.scatter(clusters, data[:,1],c=clusters,cmap=cm.brg)
plt.show()
```

```python
plt.scatter(clusters, data[:,0],c=clusters,cmap=cm.brg)
plt.show()
```
